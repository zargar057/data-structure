**2. Explain the difference between mutable and immutable data types with examples**
Mutable data types can be changed after they are created, like a list where you can add or remove items. Immutable data types cannot be changed once they are created, like a string or a number — if you want to change them, you have to make a new one. For example, lists are mutable because you can do my_list.append(4), but strings are immutable because you can’t change part of a string directly.
#my_list = [1, 2, 3]
#.append(4)  # Now it's [1, 2, 3, 4]
#list is mutable because we changed it.
#my_name = "John"
#my_name = "Jack"  # This creates a new string, it doesn't change the old one
#string is immutable because you can’t change part of it — only replace it.

#"age" in {"name": "Tom", "age": 30}  # True (checks key)

**17.In what situations are lists preferred over dictionaries?**

**18. Why are dictionaries considered unordered, and how does that affect data retrieval?**
Older versions of Python (before 3.7) did not keep dictionary items in the order they were added. Even though modern Python keeps order, the order still doesn't matter for data lookup — you find data by key, not by position.

**19. Explain the difference between a list and a dictionary in terms of data retrieval**

In a dictionary, you get data by key (like "name" or "age").

#my_dict["name"]   # by key
